```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from math import log10, sqrt
from PIL import Image



input_image = cv2.imread("/content/gdrive/MyDrive/Colab Notebooks/Imagenes_pruebas/Mandril.tiff", cv2.IMREAD_COLOR)
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(input_image, cv2.COLOR_RGB2GRAY)

plt.subplot(121)
plt.imshow(input_image)
plt.title('Imagen original')
plt.axis('off')

plt.subplot(122)
plt.imshow(image_gray, cmap = 'gray')
plt.title('Imagen en escala de gris')
plt.axis('off')
```

```python
hist1 = cv2.calcHist([image_gray],[0],None,[256],[0,256])

img_2 = cv2.equalizeHist(image_gray)
hist2 = cv2.calcHist([img_2],[0],None,[256],[0,256])

#imagen con CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
cl_img = clahe.apply(image_gray)
hist3 = cv2.calcHist([cl_img],[0],None,[256],[0,256])

#imagen original con histograma
plt.subplot(321),plt.imshow(image_gray, cmap='gray');
plt.axis('off')
plt.subplot(222),plt.plot(hist1);

#imagen equalizada con histograma
plt.subplot(323),plt.imshow(img_2, cmap='gray');
plt.axis('off')
plt.subplot(324),plt.plot(hist2);

#imagen equalizada con CLAHE con histograma
plt.subplot(325),plt.imshow(cl_img, cmap='gray');
plt.axis('off')
plt.subplot(326),plt.plot(hist3);




```
